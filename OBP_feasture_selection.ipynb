```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset with a rate target
df_rate = pd.read_csv('OBP_preprocess.csv')
df_rate
```

```python
#get correlations of each features in the dataframe
corrmat = df_rate.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df_rate[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

##Delete unimportant features that has low impact on the target

```python
df_rate=df_rate.drop(columns =['Date of Joining'])
df_rate=df_rate.drop(columns =['Gender'])
df_rate=df_rate.drop(columns =['Company Type'])
df_rate=df_rate.drop(columns =['WFH Setup Available'])
df_rate
```

```python
# saving the dataframes
df_rate.to_csv('OBP_feature_selection.csv')
```
